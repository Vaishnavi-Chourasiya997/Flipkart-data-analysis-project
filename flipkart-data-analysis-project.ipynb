## Flipkart_Data-Analysis_Project: 
    This dataset contains 5,000 products across categories like Electronics, Home, Fashion, and Appliances, providing insights into Flipkart’s sales trends.

#### Key Observations:
    1. Electronics dominates in pricing (highest ₹1,86,922 for laptops) and sales.
    2. Home & Fashion have high stock availability, suggesting slower turnover.
    3. Discounts vary widely (0% to 58%), with Electronics often discounted more.
    4. Ratings average ~4.0, but some products (e.g., Washing Machines at 3.1★) need quality checks.
    5.Top sellers include Flipkart Assured and RetailNet, indicating trust influences sales.

### Matplotlib Graphs :

#### 1. Plotting Graph : Discount % vs. Average Product Rating

```python
# Select specific records
#selected_df = df.loc[[0, 3, 5, 10, 50, 100]]  

selected_df = df.sample(frac=0.05)
# Sort and plot
selected_df = selected_df.sort_values('Discount (%)')

plt.figure(figsize=(10, 9))
plt.plot(selected_df['Discount (%)'], selected_df['Rating (★)'], marker='o', color='purple', label = 'Styled Line')
plt.title("Discount % vs. Average Product Rating")
plt.grid(True, linestyle='--', alpha=0.8)
plt.legend()
plt.xlabel("Discount Percentage (%)")
plt.ylabel("Average Rating (★)")
plt.show()
```

#### 2. Plotting Graph(Two Lines) : Discount vs Total Sold & Number of Buyers

```python
selected_df = df.loc[[20, 12,22,32,42,52,62,72,82,92,102,112,122,4222]]  # Pick selected rows
#selected_df = df.sample(frac=0.05)

# Sort for clean plot
selected_df = selected_df.sort_values('Discount (%)')

# Plot
plt.figure(figsize=(10, 10))

# First line
plt.plot(selected_df['Discount (%)'], selected_df['Total Sold'], 
         marker='o', label='Total Sold', color='blue')

# Second line (example: another column like "Price" or "Sales")
plt.plot(selected_df['Discount (%)'], selected_df['Number of Buyers'], 
         marker='s', label='Number of Buyers', color='green')

# Title and labels
plt.title('Styled Plot with - Two Lines : Discount vs Total Sold & Number of Buyers')
plt.xlabel('Discount (%)')
plt.ylabel('Values')
plt.grid(True, linestyle = ":", alpha = 0.6)
plt.legend()
plt.show()
```

#### 3. Scatter Plot Graph: Discount vs Rating & Price

```python
# Select specific rows
#selected_df = df.loc[[1,11,21,31,41,51,61,71,81,91,101,111,121,131,4111]]
selected_df = df.sample(frac=0.05)
plt.figure(figsize=(10, 7))

# First scatter: Rating
plt.scatter(selected_df['Discount (%)'], selected_df['Rating (★)'], 
            color='blue', label='Rating (★)', marker='o')

# Second scatter: Price
plt.scatter(selected_df['Discount (%)'], selected_df['Price (₹)'], 
            color='green', label='Price (₹)', marker='s')

# Labels, title, legend
plt.title('Scatter Plot: Discount vs Rating & Price')
plt.xlabel('Discount (%)')
plt.ylabel('Values')
plt.legend()
plt.show()
```

#### 4. Histograms Graph : Product Prices

```python
# Plot
plt.figure(figsize=(10, 6))
plt.hist(df['Price (₹)'], bins=50, color='skyblue', alpha=0.8)

# Labels
plt.title('Histogram of Product Prices')
plt.xlabel('Price (₹)')
plt.ylabel('Number of Products')
plt.show()
```

#### 5. Bars Graph : Top 8 Sellers by Number of Products

```python
# Count products per seller
seller_counts = df['Seller'].value_counts().head(5)

# Plotting
plt.figure(figsize=(12, 6))
colors = ['orange', 'green', 'blue', 'purple', 'red', 'cyan']
plt.bar(seller_counts.index, seller_counts.values, color=colors)

plt.title('Top 8 Sellers by Number of Products')
plt.xlabel('Seller')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(axis = 'y', linestyle='--', alpha=0.2)
plt.show()
```

#### 6. Bar Graph : Total Units Sold by Main Category

```python
# Total Sold per Main Category
total_sold = df.groupby('Main Category')['Total Sold'].sum()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(total_sold.index, total_sold.values, color='slateblue')

plt.title('Total Units Sold by Main Category')
plt.xlabel('Main Category')
plt.ylabel('Total Sold')
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.show()
```

#### 7. Pairplot : Key Numerical Features (KDE)

```python
# Sample 50% of the data
sample_df = df.sample(frac=0.5, random_state=42)

# Create a pairplot with KDE on diagonal
sns.pairplot(
    sample_df[["Price (₹)", "Rating (★)", "Number of Buyers", "Total Sold"]],
    diag_kind="kde")  # hist , auto
plt.suptitle("Pairplot of Key Numerical Features (KDE)", y=1.02)
plt.show()
```

#### 8. Pie Chart : Product Distribution by Main Category (50% sample)

```python
# Use 50% sample of the data
sample_df = df.sample(frac=0.5)

# Group by Main Category
category_counts = sample_df['Main Category'].value_counts()
color = ["skyblue", "orange", "violet", "pink"]

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%', colors=color)
plt.title("Product Distribution by Main Category (50% sample)",pad = 20)
plt.axis('equal')  # Equal aspect ratio makes the pie round.
plt.show()
```
